```python
X = df.drop('Profit',axis = 1)
```

# Applying the KMeans clustering to the MLR_Startups dataset

Predicted the Accuracy using rand_score

```python
KM = KMeans(n_clusters = 10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3)
```

```python
KM.fit(x_train,y_train)
```

```python
predicted_values = KM.predict(x_test)
```

```python
from sklearn.metrics import rand_score 
```

```python
print("accuracy score of Kmeans clustering :",rand_score(y_test,predicted_values))
```
